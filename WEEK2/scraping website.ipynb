```python
import os
import pandas as pd
import requests
from bs4 import BeautifulSoup

```

```python
WEB = "http://books.toscrape.com/"
html_script = requests.get(WEB).text
```

```python
soup = BeautifulSoup(html_script,'html.parser')
```

```python
genre = soup.find("ul",class_="nav nav-list")
genre_list = genre.find_all("a")
print(genre_list)
```

```python
cat_name=[]
cat_link=[]

for lst in genre_list:
    cat_name.append(lst.text.strip())
    cat_link.append(lst["href"])
    
cat_name.pop(0)
cat_link.pop(0)
```

```python
df = pd.DataFrame({"Category_name" : cat_name,"Category_links" : cat_link })
df.head()
```

```python
df.to_csv(os.path.join("scraped_genres.csv"))
```
