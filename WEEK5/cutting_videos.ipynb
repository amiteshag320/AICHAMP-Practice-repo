```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
cap = cv.VideoCapture('cricket.mp4')
total = cap.get(7)
cap.set(1,9025)
ret,frame = cap.read()
cv.imwrite("cric.jpeg",frame)

cap.set(1,1318)
ret,frame = cap.read()
cv.imwrite("cric1.jpeg",frame)

cap.set(1,245)
ret,frame = cap.read()
cv.imwrite("cric2.jpeg",frame)

cap.set(1,1666)
ret,frame = cap.read()
cv.imwrite("cric3.jpeg",frame)
```

```python
img=cv.imread("cric.jpeg")
img = cv.cvtColor(img ,cv.COLOR_BGR2GRAY)
count=0
hist = cv.calcHist([img],[0],None,[256],[0,256])
img = img.reshape(225280,1)
for i in img:
    if 200<i:
        count+=1
plt.plot(hist)
count
```

```python
img1=cv.imread("cric1.jpeg")
img1 = cv.cvtColor(img1 ,cv.COLOR_BGR2GRAY)
count=0
hist_2 = cv.calcHist([img1],[0],None,[256],[0,256])
img1 = img1.reshape(225280,1)
for i in img1:
    if 100<i<150:
        count+=1
plt.plot(hist_2)

cv.compareHist(hist, hist_2,0)
count
```
