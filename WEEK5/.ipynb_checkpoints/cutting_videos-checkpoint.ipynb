```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
cap = cv.VideoCapture('cricket.mp4')
total = cap.get(7)
cap.set(1,175)
ret,frame = cap.read()
cv.imwrite("cric.jpeg",frame)

cap.set(1,75)
ret,frame = cap.read()
cv.imwrite("cric1.jpeg",frame)
cap.release()
```

```python
img=cv.imread("cric.jpeg")
img = cv.cvtColor(img ,cv.COLOR_BGR2GRAY)
img = cv.GaussianBlur(img , (5,5),0)
hist = cv.calcHist([img],[0],None,[256],[0,256])
plt.plot(hist)
```

```python
img1=cv.imread("cric1.jpeg")
img1 = cv.cvtColor(img1 ,cv.COLOR_BGR2GRAY)
img = cv.GaussianBlur(img , (5,5),0)

hist_2 = cv.calcHist([img1],[0],None,[256],[0,256])
plt.plot(hist_2)

cv.compareHist(hist, hist_2,3)

```

```python
cap = cv.VideoCapture('cricket.mp4')
i=0
while cap.isOpened():
    ret,frame = cap.read()
    gray = cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
    img = cv.GaussianBlur(gray , (11,11),0)
    histog =  cv.calcHist([img],[0],None,[256],[0,256])
    
    thresh = cv.compareHist(hist, histog,3)
    if thresh < 0.35:
        cv.imwrite("frames/frame{}.jpeg".format(i),frame)
        i+=1
    #if i==700:
     #   break
        
```
